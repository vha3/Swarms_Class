```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Meetings between Sprites were computed using a distance threshold
# of 0.2km. These simulations each included 100 Sprites over 60sec
# timesteps.
# meetings.csv has one column per Sprite and each row a timestep.
meetings = np.genfromtxt('meetings.csv', delimiter=',')
```

```python
def plot_meetings(meetings, title):
    plt.xlabel('Timestep (minutes)')
    plt.ylabel('Encounters')
    plt.title(title)
    plt.plot(meetings)
    plt.show()

# Early period cut off point in minutes
early_stop = 30

# Create two plots:
# First 30 minutes (all Sprites / average Sprite / one Sprite)
plot_meetings(meetings[:early_stop,:], 'Pairwise spacecraft encounters in the first 30 minutes')

# After first 30 minutes (all Sprites / average Sprite / one Sprite)
plot_meetings(meetings[early_stop:,:], 'Pairwise spacecraft encounters after the first 30 minutes')
```
